```python
# --- 1. THE POWER-UP (Installation) ---
!pip install GoogleNews yfinance textblob

import yfinance as yf
from GoogleNews import GoogleNews
from textblob import TextBlob
import logging
from datetime import datetime

# --- 2. THE HEALY VECTOR LABS CORE ---
def veritas_audit(sentiment_score, transit_factor):
    """VERITAS: The Truth Multiplier."""
    return round(sentiment_score * transit_factor, 2)

def get_live_sentiment(ticker):
    """The 'Noise' Sensor with Bulletproof Error Handling."""
    try:
        gn = GoogleNews(period='1d', lang='en')
        gn.search(f'{ticker} ceasefire')
        results = gn.result()

        # FIX: Check if results is None or Empty before iterating
        if not results:
            print(f"!!! LAB NOTE: No live news found for {ticker}. Defaulting to neutral (0.5).")
            return 0.5

        headlines = [res.get('title', '') for res in results if res and res.get('title')]

        if not headlines:
            return 0.5

        scores = [TextBlob(h).sentiment.polarity for h in headlines]
        avg_score = sum(scores) / len(scores)
        return round((avg_score + 1) / 2, 2)
    except Exception as e:
        print(f"!!! SCRAPER ALERT: {e}. Defaulting to 0.5.")
        return 0.5

# --- 3. MASTER EXECUTION ---
ticker = "XOM"

try:
    # A. Market Data Check
    stock = yf.Ticker(ticker)
    # Using 'fast_info' can be buggy; history is the 'Senior' move for stability
    hist = stock.history(period='1d')
    current_price = hist['Close'].iloc[-1] if not hist.empty else 0.0

    sentiment_reading = get_live_sentiment(ticker)

    # B. The Physical Truth (April 8, 2026 Strategy)
    # Market says "Peace" (Sentiment High), but only 11-15 ships moved today.
    # Baseline 135/day. VERITAS Factor = 15/135 = 0.11
    physical_truth_multiplier = 0.11

    # C. Calculate Kinetic Alpha & FASE Score
    # FASE = (Fundamentals 0.65) + (Audited Sentiment 0.35)
    veritas_score = veritas_audit(sentiment_reading, physical_truth_multiplier)
    fase_score = (0.65) + (veritas_score * 0.35)

    # D. THE HEALY VECTOR LABS DASHBOARD
    print(f"\n" + "█"*45)
    print(f"   HEALY VECTOR LABS | FASE Master v3.2")
    print(f"   TIMESTAMP: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
    print("█"*45)
    print(f"TICKER: {ticker} | LIVE PRICE: ${round(current_price, 2)}")
    print("-" * 45)
    print(f"NEWS SENTIMENT (Noise):       {sentiment_reading}")
    print(f"VERITAS (Truth Multiplier):   {veritas_score}")
    print(f"FINAL FASE MASTER SCORE:      {round(fase_score, 2)}")
    print("-" * 45)

    if veritas_score < 0.20:
        print("ACTION: STRONG BULLISH DIVERGENCE")
        print("REASON: Geopolitical noise is fake. Physical supply is still choked. BUY.")
    else:
        print("ACTION: MARKET SYNC")
        print("REASON: Truth matches the Noise.")
    print("█"*45)

except Exception as e:
    print(f"!!! CRITICAL LAB ERROR: {e}")
```
